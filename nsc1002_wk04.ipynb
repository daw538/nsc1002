## Algorithms
An algorithm is a process, set of instructions, or set of rules used in a computation. In the last workshop we experimented with the method of Heron of Alexandria and Exhaustive Enumeration. In the next sections we will explore two additional algorithms to advance our experience with loops, conditionals, variables, and lists.

### Root finding algorithms
The idea of root finding algorithms is to solve the equation:

$$ F\left(x\right) = 0.$$

In our example, if we are interested in finding the square root of a number $x_0$ with the answer being $x$ then we are interested in,

$$ x_0 = \sqrt{x} $$

and thus the equation we're trying to solve with a root solver is

$$ \sqrt{x} - x_0 = 0, $$

therefore

$$ F\left(x\right) = \sqrt{y}-x=0.$$

Or, if we square each side of the initial equation we can alternatively write this,

$$ F\left(x\right) = y-x^2 = 0. $$

Refering to the initial equation, we can test this equation by looking at a familiar example: in the case where we are searching for the square root of 4, $y$ is 4 and $x_0$ is 2. If we were searching for the square root of 5, then $y = 5$ and $x_0$ is the unknown number we're going to compute with an algorithm.

#### 1. Exhaustive Enumeration (reminder) or “brute force search”
This is a general method of finding the solution to a problem. We begin with a guess, change it by a small amount, check, and continue until it's close enough. Here is demonstration code from the class reference book _(Introduction to Computation and Programming using Python, by John Guttage, Figure 3.3)_ that finds the square root of $x$:
```python
"""Find square root of number using exhaustive enumeration."""
y = 25.0
epsilon = 0.01
step = epsilon**2
num_guesses = 0
tot_guesses = 1.e7
ans = 0.0

while abs(ans**2 - y) >= epsilon and num_guesses <= tot_guesses:
    ans += step
    num_guesses += 1
    
print('Num Guesses: {}'.format(num_guesses))
if abs(ans**2 - y) >= epsilon:
    print(f'Failed to find sqrt of {x}')
else:
    print(f'{ans} is close to sqrt of {y}')
```

#### 2. Bisection Search
This is a root-finding method that repeatedly bisects an interval, then chooses which subinterval contains the root before continuing the search. The method is often robust, but also not as efficient as some other methods.

The following code is an algorithm to approximate the square root of $x$ using the Bisection Method, from Guttag's book.
```python
"""Find square root of number using bisection."""
y = 25
eps = .01
num_guesses = 0
tot_guesses = 1e2
low = 0.0
high = max(1.0, y)
ans = (high + low)/2.0

while abs(ans**2 - y) >= eps and num_guesses < tot_guesses:
    print('L= {} H= {} ANS= {}'.format(low, high, ans))
    num_guesses += 1
    if ans**2 < y:
        low = ans
    else:
        high = ans
        ans = (high + low) / 2.0
print(f'num_guesses = {num_guesses}')
print(f'{ans} is close to square root of {x}')
```

#### 3. Newton-Raphson
A an approximation algorithm attributed to Isaac Newton and Joseph Raphson. It is often simply called Newton's Method. It can be used to find the roots of many different types of functions. For us we will use it only for our polynomial problem. First, to more easily understand the Newton Raphson algorithm, we consider an alternative way of thinking about finding the root than we did in with equation from the Bisection method. Consider

$$ F\left(y\right) = 0, \qquad \textrm{y is a real number.}$$

Define the derivative of $F$ to be $F^\prime = \frac{\textrm{d}F}{\textrm{d}y}$. If we begin with a guess $y_0$ then a good second guess, according to the Newton Raphson method is,

$$ y1 = y0 − \frac{F\left(y_0\right)}{F^\prime\left(y_0\right)}. $$

This process is then repeated. For programs to find the square root we are considering the polynomial $F\left(y\right) = y \times y − x = 0$, where $x$ represents the number we want to know the square root of, so it is fixed. Taking the derivative we find that $F^\prime\left(y\right) = 2y$. So for the algorithm of interest the new guess for the iteration
$n+1$ can be written:

$$ y_{n+1} = y_n - \frac{{y_n}^2 -x}{2y_n} $$

The following is an algorithm to approximate the square root of $x$ using the Newton–Raphson method from our class text book Figure 3.5, _"Introduction to Computation and Programming Using Python"_ by John V. Guttag.

```python
"""Find square root of number using Newton-Raphson."""
y = 25.0
epsilon = .01
guess = y / 2.0 # first guess

while abs(guess * guess - y) >= epsilon:
    guess -= (guess**2 - y) / (2 * guess)
print('{} is close to square of {}'.format(guess, y))
```
It turns out that this algorithm is exactly the same as the one due to Heron of Alexandria we used in the last Workshop. Can you see why? Strategies for developing an algorithm.


## Strategies for developing an algorithm
For your assessment you’ll be asked to write an algorithm that solves a problem using python. These are a few ideas you can use to develop an algorithm
- Use a flow chart – see Figure 1.

<img src="FlowChartFunction.png" alt="Figure 2: Flow Chart Example: Branching can be seen easily with if statements. Instructions for actions in boxes." width="600"/>

1. Identify the main processes (or calculations):
2. Identify the main decision points (or if/elif/else)
3. Identify possible loops (or iterations)
4. Identify the variables that you know. These may need to be initialized.
5. Identify the outputs that you want.
6. Sketch the flow of the algorithm by using a flow chart. One of the best ways to get started making a flow chart is to write what you know at the top of a piece of paper (these variables may need to be initialized) and what you want at the bottom. Fill in the required steps and conditionals to get from what you know to what you need. 
7. Step through what you have to see if it makes sense.

- The creative drawing method can help. Begin by writing down the processes, conditional tests, data that you know, the output or endpoint on a piece of paper. Circle each step. Draw circles around groups of processes that are connected. This is illustrated on the left side of the figure below. Use lines and arrows to show yourself where they fit into the 'flow' of the algorithm. Once you have a good idea, redraw the flow. This is illustrated on the right side of the figure. Do this until you have sketched the flow of the algorithm. There are many ways of developing and writing the same algorithm, be creative in how you think about it.

The sketch on the right is a refinement of the one on the left. The idea is to modify the sketch until you can clearly see each step of the algorithm.
